```python
def fetch_stock_data(stock_ticker, num_of_days=59, interval = "15m"):
    import yfinance as yf
    import pandas as pd
    import numpy as np
    import datetime as dt
    import pprint


    stock_ticker = stock_ticker  # Replace with your desired stock ticker
    start_date = (dt.datetime.now() - dt.timedelta(days=num_of_days)).strftime('%Y-%m-%d')
    end_date = dt.datetime.now().strftime('%Y-%m-%d')

    # Fetch historical data
    data = yf.download(stock_ticker, start=start_date, end=end_date, interval= interval)

    return data # Return the data





# Call the function and display the data
data = fetch_stock_data(stock_ticker="^NSEBANK", num_of_days=59, interval = "15m")
print(type(data))
data.to_csv('BANKNIFTY_stock_data.csv', index=True)
display(data) # Display the data once
```

```python
import pandas as pd

def read_and_preprocess_stockData(file_path):
    """
    Reads stock data from a CSV file, performs preprocessing steps,
    and returns the processed DataFrame.

    Args:
        file_path (str): The path to the CSV file.

    Returns:
        pandas.DataFrame: The processed DataFrame.
    """

    import pandas as pd
    # Read stock data from csv
    df = pd.read_csv(file_path)

    # Drop first two rows and reset index
    df = df.drop([0, 1])
    df = df.reset_index(drop=True)

    # Rename 'Price' column to 'Datetime'
    df = df.rename(columns={"Price": "Datetime"})

    # Convert columns to numeric and round to 2 decimal places
    for column in df.columns:
        if column != "Datetime":
            df[column] = pd.to_numeric(df[column], errors='coerce').round(2)

        # Set 'Datetime' column as index and convert it to DatetimeIndex
    df = df.set_index('Datetime')
    df.index = pd.to_datetime(df.index)

    return df

import pandas_ta as ta

class TechnicalIndicators:
    """
    A class for calculating technical indicators using pandas_ta
    and adding them as new columns to the DataFrame.
    """

    def __init__(self, data):
        """
        Initializes the TechnicalIndicators class with the input data.

        Args:
            data (pandas.DataFrame): The input DataFrame containing OHLCV data.
        """
        self.data = data

    def calculate_ema(self, length=14, column_name="EMA"):
        """
        Calculates the Exponential Moving Average (EMA) and adds it as a new column.

        Args:
            length (int, optional): The length of the EMA window. Defaults to 14.
            column_name (str, optional): The name of the new column for EMA. Defaults to "EMA".
        """
        self.data[column_name+"_"+str(length)] = ta.ema(self.data["Close"], length=length)

    def calculate_sma(self, length=14, column_name="SMA"):
        """
        Calculates the Simple Moving Average (SMA) and adds it as a new column.

        Args:
            length (int, optional): The length of the SMA window. Defaults to 14.
            column_name (str, optional): The name of the new column for SMA. Defaults to "SMA".
        """
        self.data[column_name+"_"+str(length)] = ta.sma(self.data["Close"], length=length)

    def calculate_rsi(self, length=14, column_name="RSI"):
        """
        Calculates the Relative Strength Index (RSI) and adds it as a new column.

        Args:
            length (int, optional): The length of the RSI window. Defaults to 14.
            column_name (str, optional): The name of the new column for RSI. Defaults to "RSI".
        """
        self.data[column_name+"_"+str(length)] = ta.rsi(self.data["Close"], length=length)

    def calculate_vwap(self, column_name="VWAP"):
        """
        Calculates the Volume Weighted Average Price (VWAP) and adds it as a new column.

        Args:
            column_name (str, optional): The name of the new column for VWAP. Defaults to "VWAP".
        """
        self.data[column_name] = ta.vwap(self.data["High"], self.data["Low"], self.data["Close"], self.data["Volume"])
```

```python
tcs_data = read_and_preprocess_stockData('/content/BANKNIFTY_stock_data.csv')
display(tcs_data)
# Assuming you have a DataFrame named 'df' with OHLCV data
indicators = TechnicalIndicators(tcs_data)

# Calculate indicators and add columns
indicators.calculate_ema(length=9, column_name="EMA")
indicators.calculate_sma(length=5, column_name="SMA")
indicators.calculate_rsi(length=14, column_name="RSI")
indicators.calculate_vwap(column_name="VWAP")

# The DataFrame 'df' now has the new columns added
display(tcs_data.head(30))
```

```python

import pandas as pd
import pandas_ta as ta

def generate_signals(data, indicators, entry_conditions):
    """
    Generates trading signals based on specified indicators and entry conditions.

    Args:
        data (pandas.DataFrame): The DataFrame containing OHLCV data and indicators.
        indicators (list): A list of indicator column names to consider.
        entry_conditions (list): A list of entry conditions as strings.
                                  Each condition should be a comparison
                                  involving an indicator and the closing price.

    Returns:
        pandas.DataFrame: The DataFrame with an added 'Signal' column
                          indicating entry points (1) or no entry (0).
    """

    # Initialize the 'Signal' column to 0
    data['Signal'] = 0

    # Evaluate entry conditions and set 'Signal' to 1 where they are met
    for condition in entry_conditions:
        # Example condition: "Close > VWAP and RSI > 60"
        data.loc[data.eval(condition), 'Signal'] = 1

    return data

def populate_entry_exit(data, stock_name, entry_conditions, exit_conditions):
    """
    Populates a new DataFrame with stock name, entry price, exit price, and datetime.

    Args:
        data (pandas.DataFrame): The DataFrame with signals.
        stock_name (str): The name of the stock.
        entry_conditions (list): A list of entry conditions as strings.
        exit_conditions (list): A list of exit conditions as strings.

    Returns:
        pandas.DataFrame: A DataFrame with stock name, entry price, exit price, and datetime.
    """

    entry_exit_data = []
    entry_price = None
    entry_datetime = None

    for index, row in data.iterrows():
        # Use data.eval() directly with the index for entry conditions
        if data.eval(entry_conditions[0])[index] and entry_price is None:  # Entry condition met
            entry_price = row['Close']
            entry_datetime = row.name  # index is the DatetimeIndex
        elif entry_price is not None:  # Check exit conditions if in a trade
            # Check if any of the exit conditions are met, using the index directly
            if any(data.eval(condition)[index] for condition in exit_conditions):
                entry_exit_data.append([stock_name, entry_price, row['Close'], entry_datetime, row.name, row["Close"] - entry_price])
                entry_price = None
                entry_datetime = None

    return pd.DataFrame(entry_exit_data, columns=['Stock Name', 'Entry Price', 'Exit Price', 'Entry Datetime', 'Exit Datetime', 'P&L'])

# Define your entry and exit conditions
indicators = ['VWAP', 'RSI_14', 'EMA_9']
entry_conditions = ['RSI_14 > 60 and Close > EMA_9']
exit_conditions = ['RSI_14 < 50 and EMA_9<SMA_5']  # Your exit conditions

# Generate signals (you might already have this part)
tcs_data = generate_signals(tcs_data, indicators, entry_conditions)

# Populate entry/exit DataFrame with the specified entry and exit conditions
entry_exit_df = populate_entry_exit(tcs_data, 'TCS.NS', entry_conditions, exit_conditions)

display(tcs_data)
#
# Display the entry/exit DataFrame
display(entry_exit_df)
print(sum(entry_exit_df["P&L"]))
```

```python
import pandas as pd

def calculate_drawdown(df):
    """
    Calculates the drawdown of a trading strategy based on the P&L column.

    Args:
        df (pandas.DataFrame): DataFrame containing the 'P&L' column.

    Returns:
        pandas.DataFrame: DataFrame with an additional 'Drawdown' column.
    """

    # Calculate cumulative P&L
    df['Cumulative_PnL'] = df['P&L'].cumsum()

    # Calculate peak P&L
    df['Peak_PnL'] = df['Cumulative_PnL'].cummax()

    # Calculate drawdown
    df['Drawdown'] = (df['Peak_PnL'] - df['Cumulative_PnL'])

    return df
```
